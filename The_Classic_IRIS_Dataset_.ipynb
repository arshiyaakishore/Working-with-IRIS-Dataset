```python
col='sepal_length'
sns.relplot(x=col,y="species",hue="species",data=data)
plt.title("Sepal_length vs Species")
```

```python
col='sepal_width'
sns.relplot(x=col,y="species",hue="species",data=data)
plt.title("Sepal_width vs Species")
```

In the above graph between Sepal width and Species we can not distinguish any flower as the graph has overlapping

```python
col='petal_length'
sns.relplot(x=col,y="species",hue="species",data=data)
plt.title("Petal_length vs Species")
# plt.grid()
```

Here in the Above graph we can clearly distinguish Setosa flower on the basis of the Petal Length

```python
col='petal_width'
sns.relplot(x=col,y="species",hue="species",data=data)
plt.title("Petal_width vs Species")
```

Here in the Above graph we can clearly distinguish Setosa flower on the basis of the Petal width too
